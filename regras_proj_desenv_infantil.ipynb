```python
#Gera fatos dos dados das crianças
def gera_fatos_criancas(lista_criancas):
  fatos_criancas = []
  for c in lista_criancas:
    fato = f"crianca({c['nome']}_{c['id_crianca']}, {c['idade_meses']})"
    fatos_criancas.append(fato)
  return fatos_criancas
```

```python
#Gera fatos dos marcos
def gera_fatos_marcos(crianca_filtrada_idade_avaliacao):
    fatos_marco_desenvolvimento = [] #lista para armazenar a string
    marcos_lista = []
    registro_idade_no_marco = []
    limite_superior_de_idade_do_marco = []

    marcos_crianca = crianca_filtrada_idade_avaliacao['marcos']
    for mrc_crianca in marcos_crianca:
        fato_marco_crianca = list(mrc_crianca.keys())[0]
        marco = fato_marco_crianca.split(')')[0]
        fato_marco = f"{marco}({crianca_filtrada_idade_avaliacao['nome']}_{crianca_filtrada_idade_avaliacao['id_crianca']}, {mrc_crianca[fato_marco_crianca]['avaliacao']}, {mrc_crianca[fato_marco_crianca]['registro_idade_mes']}, {mrc_crianca[fato_marco_crianca]['limite_superior_faixa_etaria']})"
        fatos_marco_desenvolvimento.append(fato_marco)
        marcos_lista.append(marco)
        registro_idade_no_marco.append(mrc_crianca[fato_marco_crianca]['registro_idade_mes'])
        limite_superior_de_idade_do_marco.append(mrc_crianca[fato_marco_crianca]['limite_superior_faixa_etaria'])

    return fatos_marco_desenvolvimento, marcos_lista,registro_idade_no_marco,limite_superior_de_idade_do_marco

# fatos_marco_desenvolvimento, marcos_lista = gera_fatos_marcos(crianca)
```

```python
#Gera Regras dos marcos
def gera_regras_marcos(marcos_lista):
  regras_marcos = []
  regras_marcos_lista = list(OrderedDict.fromkeys(marcos_lista)) #retira duplicadas

  for rgr in regras_marcos_lista:
    regra_txt = f"{rgr}_presente(C) :- {rgr}(C ,presente, Reg_idade_marco, Limite_superi_faixa),crianca(C, Idade), (Reg_idade_marco <= Limite_superi_faixa and Idade >= Reg_idade_marco)"
    regras_marcos.append(regra_txt)
  return regras_marcos
```

### Aplicação de consultas sobre a base de conhecimento

```python
def consulta_marcos_presentes_para_crianca(crianca_nome,marcos_lista,base_conhecimento):
  lista_retorno_consulta = []

  for m in marcos_lista:
    consulta = f"{m}_presente({crianca_nome})"
    resultado_consulta = base_conhecimento.query(pl.Expr(consulta))
    lista_retorno_consulta.append(resultado_consulta[0])

  return lista_retorno_consulta

def dev_adequado_crianca(crianca_nome,marcos_lista,base_conhecimento):
  resultados_marcos_lista = consulta_marcos_presentes_para_crianca(crianca_nome, marcos_lista,base_conhecimento)

  classificacao = False
  for r_mrc in resultados_marcos_lista:
    if r_mrc == "No":
      classificacao = False
      break
    else:
      classificacao = True

  return classificacao
```

###Identificação de marcos ausentes, atrasos, adequados e em alertas

```python
def identificar_marcos_ausentes(resultados_marcos_lista,marcos_lista):
  marcos_ausentes = []
  for i in range(len(resultados_marcos_lista)):
    if resultados_marcos_lista[i] == 'No':
      marcos_ausentes.append(marcos_lista[i])
  return marcos_ausentes
```

```python
def identificar_indicacoes_alertas_atrasos(crianca, marcos_ausentes):
  marcos_crianca_selecionada = crianca['marcos']
  #print(list(marcos_crianca_selecionada[1].keys())[0])
  idade_crianca = int(crianca['idade_meses'])

  lista_indicacoes_marcos_alertas_atrasos = []
  for i in range(len(marcos_crianca_selecionada)):
    for j in range(len(marcos_ausentes)):
      if list(marcos_crianca_selecionada[i].keys())[0] == marcos_ausentes[j]:
        registro_idade_mes = marcos_crianca_selecionada[i][marcos_ausentes[j]]['registro_idade_mes']
        limite_superior_faixa_etaria = marcos_crianca_selecionada[i][marcos_ausentes[j]]['limite_superior_faixa_etaria']
        limite_inferior_faixa_etaria = marcos_crianca_selecionada[i][marcos_ausentes[j]]['limite_inferior_faixa_etaria']


        if (idade_crianca <= 1):
          lista_indicacoes_marcos_alertas_atrasos.append('atraso')

        else:
          if (idade_crianca >= int(limite_inferior_faixa_etaria) and idade_crianca <= int(limite_superior_faixa_etaria)):
            lista_indicacoes_marcos_alertas_atrasos.append('alerta')

          else:
            lista_indicacoes_marcos_alertas_atrasos.append('atraso')
  return lista_indicacoes_marcos_alertas_atrasos
```

```python
def gerar_indicacoes_ad_at_al_marcos(resultados_marcos_lista, lista_indicacoes_marcos_alertas_atrasos):
  lista_indicacoes_marcos_adequado_alertas_atrasos = []

  if not lista_indicacoes_marcos_alertas_atrasos:
    for r_m in resultados_marcos_lista:
      lista_indicacoes_marcos_adequado_alertas_atrasos.append('AD')
  else:
    for r_m in resultados_marcos_lista:
      if r_m == 'No':

        if lista_indicacoes_marcos_alertas_atrasos[0] == 'atraso':
          abr_indicacao = 'AT'
        if lista_indicacoes_marcos_alertas_atrasos[0] == 'alerta':
          abr_indicacao = 'AL'

        lista_indicacoes_marcos_adequado_alertas_atrasos.append(abr_indicacao)
        lista_indicacoes_marcos_alertas_atrasos.pop(0)
      else:
        lista_indicacoes_marcos_adequado_alertas_atrasos.append('AD')

  return lista_indicacoes_marcos_adequado_alertas_atrasos
```

```python
def filtrar_faixa_etaria_anterior_atual(crianca):
    lista_crianca_filtrada_marcos_atual = []

    crianca_filtrada = crianca.copy()
    idade_meses = int(crianca_filtrada['idade_meses'])

    marcos_adicionados = set()  # Conjunto para armazenar os marcos já adicionados

    for marco in crianca_filtrada['marcos']:
        chave_marco = list(marco.keys())[0]
        limite_superior = int(marco[chave_marco]['limite_superior_faixa_etaria'])
        limite_inferior = int(marco[chave_marco]['limite_inferior_faixa_etaria'])

        # Faixa etária atual
        if limite_inferior <= idade_meses <= limite_superior:
            lista_crianca_filtrada_marcos_atual.append(marco)

            # Adiciona os marcos da faixa etária anterior
            limite_inferior_anterior = limite_inferior - 1
            for marco_anterior in crianca_filtrada['marcos']:
                chave_marco_anterior = list(marco_anterior.keys())[0]
                limite_superior_anterior = int(marco_anterior[chave_marco_anterior]['limite_superior_faixa_etaria'])
                limite_inferior_anterior_marco = int(marco_anterior[chave_marco_anterior]['limite_inferior_faixa_etaria'])

                if limite_inferior_anterior_marco <= limite_inferior_anterior <= limite_superior_anterior:
                    chave_marco_anterior = list(marco_anterior.keys())[0]
                    if chave_marco_anterior not in marcos_adicionados:  # Verifica se o marco já foi adicionado
                        lista_crianca_filtrada_marcos_atual.append(marco_anterior)
                        marcos_adicionados.add(chave_marco_anterior)  # Adiciona o marco ao conjunto de marcos já adicionados

    crianca_filtrada['marcos'] = lista_crianca_filtrada_marcos_atual
    return crianca_filtrada
```
